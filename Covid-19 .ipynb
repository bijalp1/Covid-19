# Total number of Male Vs. Female

```python

df = pd.read_csv('/Users/pratikzaveri/Desktop/Dataset/Covid_Data.csv')

# Count for total number of females (where gender is 1)
total_females = (df['SEX'] == 1).sum()

# Count for total number of males (where gender is 2)
total_males = (df['SEX'] == 2).sum()

# Display the results
print(f'Total Females: {total_females}')
print(f'Total Males: {total_males}')

# Bar graph
plt.bar(['Female', 'Male'], [total_females, total_males], color=['pink', 'blue'])
plt.xlabel('SEX')
plt.ylabel('Total Count')
plt.title('Total Number of Males and Females in the Dataset')
plt.text(0, total_females + 5, str(total_females), ha='center', va='bottom')
plt.text(1, total_males + 5, str(total_males), ha='center', va='bottom')

plt.show()
```

# Total Patients Alive vs. Dead

```python
df = pd.read_csv('/Users/pratikzaveri/Desktop/Dataset/Covid_Data.csv')

# Count for total number of patients (alive as well as died)
total_counts = df['DATE_DIED'].value_counts()

# Count for total number of patients alive (where date died = 9999-99-99)
Alive = (df['DATE_DIED'] == '9999-99-99').sum()

# Count for total number of patients dead (where date died != 9999-99-99)
Died = (df['DATE_DIED'] != '9999-99-99' ).sum()


print(f'Total Number of Patients Died: {Died}')
print(f'Total Number of patients Alive: {Alive}')

# Bar graph
plt.bar(['Alive','Died'], [alive, died], color =['purple','pink'])
plt.xlabel('Patient status')
plt.ylabel('Number of Patients')
plt.title('Total number of Patients Died vs. Alive')
plt.text(0, Alive + 5, str(Alive), ha='center', va='bottom')
plt.text(1, Died + 5, str(Died), ha='center', va='bottom')
plt.show()
```

# Total Patient Returned Home vs. Hospitalized

```python
df = pd.read_csv('/Users/pratikzaveri/Desktop/Dataset/Covid_Data.csv')

# Count the number of patients with status 1 (returned home) and status 2 (hospitalized)
status_counts = df['PATIENT_TYPE'].value_counts()

# Count for total number of patients returned home (where Patient type is 1)
returned_home = (df['PATIENT_TYPE'] == 1).sum()

# Count for total number of patients hospitalized (where patient type is 2)
hospitalized = (df['PATIENT_TYPE'] == 2).sum()

# Display the results
print(f'Total Patients returned home: {returned_home}')
print(f'Total Patients Hospitalized: {hospitalized}')

# Plot the results
plt.bar(status_counts.index, status_counts.values, color=['green', 'red'])
plt.xlabel('Patient Type')
plt.ylabel('Number of Patients')
plt.title('Total Number of Patients Returned Home vs. Hospitalized')
plt.text(1, returned_home + 5, str(returned_home), ha='center', va='bottom')
plt.text(2, hospitalized + 5, str(hospitalized), ha='center', va='bottom')
plt.xticks(status_counts.index, ['Returned Home', 'Hospitalized'])
plt.show()
```

# Patients with Underlying medical Condition (medthod 2)

```python
df = pd.read_csv('/Users/pratikzaveri/Desktop/Dataset/Covid_Data.csv')

# List of underlying medical conditions
conditions = ['ASTHMA', 'DIABETES', 'HIPERTENSION', 'CARDIOVASCULAR', 'PNEUMONIA', 'OBESITY', 'INMSUPR', 'RENAL_CHRONIC', 'TOBACCO']

# Count for total number of patients for each condition
counts = [(df[condition] == 1).sum() for condition in conditions]
plt.figure(figsize=(10, 6))
# Bar graph
plt.bar(conditions, counts, color=['orange', 'yellow', 'red', 'blue', 'purple', 'green', 'skyblue', 'pink', 'magenta'])

# Text annotations for each bar
for i, count in enumerate(counts):
    plt.text(i, count + 5, str(count), ha='center', va='bottom')

plt.xlabel('Underlying Medical Condition')
plt.ylabel('Number of Patients')
plt.title('Disease')
plt.show()
```
